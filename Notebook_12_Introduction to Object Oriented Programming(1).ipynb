### Add some attributes and methods

```python
# Create an Option class that does nothing (yet)
class Circle(object):
    
    # special method to initialise data members - always use self for the first argument
    def __init__(self,radius=0):
        self.radius = radius
```

```python
# The self object is added in automatically
b = Circle(radius=2.0)
```

```python
b
```

```python
# This is an attribute - not a method/function
b.radius
```

Let's fill out the class 

```python
import numpy as np
# Create an Option class that does nothing (yet)
class Circle(object):
        
    # special method to initialise data members - always use self for the first argument
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        area = np.pi * self.radius * self.radius
        return area
```

```python
# The self object is added in automatically

b = Circle(radius= 2.0)
b.area()
```

We can change the radius explicitly

```python
b.radius = 3.0
```

```python
b.area()
```

## Hiding Data Members

```python
import numpy as np

class Circle(object):
        
    def __init__(self,x):
        self._radius = x
        
    def area(self):
        area = np.pi * self._radius * self._radius
        return area
    
    def setRadius(self, r):
          self._radius = r
```

```python
b=Circle(1.2)
```

```python
b.area()
```

```python
b.setRadius(2.0)
```

```python
b.area()
```
