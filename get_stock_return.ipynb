## 思路:

### 先直接在完整的表中算60天return
### 然后把df4中的股票选出来
### 然后直接dropna (我觉得相当于把时间不够长的筛掉了)
### 检查一下是否在df4的时间范围内

```python
# 选择df4中的所有股票
df4_tickers = df4.SECUCODE.unique()
```

```python
# 根据定义的时间计算return
time_gap = 6 * 30
df3['RET'] = df3.groupby('SECUCODE')['S_DQ_ADJCLOSE'].transform(lambda x: x.pct_change(time_gap))
```

```python
df4_with_RET = (df3.set_index('SECUCODE')
                   .loc[df4_tickers] # 将需要的股票筛选出来
                   .reset_index()
                   .pipe(lambda df: df[df.RET.notnull()]) # 把return为nan的都去掉
               ) 
```

```python
df4_with_RET.sort_values('TRADEDATE') # 看一下时间是否满足，这里恰巧满足df4的时间（从2013-08-12开始）
```
