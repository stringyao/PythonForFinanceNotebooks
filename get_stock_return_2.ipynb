```python
import numpy as np
import pandas as pd 
```

```python
df3 = pd.read_csv('../input/jt_data/oneyear.csv',index_col=0)
df4 = pd.read_csv('../input/jt_data/threemonth.csv',index_col=0)
```

```python
df3.TRADEDATE = pd.to_datetime(df3.TRADEDATE) # 处理一下时间
df3 = df3.sort_values(['SECUCODE', 'TRADEDATE']) # 进行排序，保证时间顺序
df3 = df3.set_index('TRADEDATE')

df4.TRADEDATE = pd.to_datetime(df4.TRADEDATE) # 处理一下时间
df4 = df4.sort_values(['SECUCODE', 'TRADEDATE']) # 进行排序，保证时间顺序
```

```python
def get_return(row):
    
    time_gap = pd.Timedelta(days=6 * 30)
    
    currentdate = row.TRADEDATE
    ticker = row.SECUCODE
    currentPx = row.S_DQ_ADJCLOSE
    
    stockSlice = df3[df3.SECUCODE == ticker].copy()
    stockSlice = stockSlice.loc[:currentdate - time_gap] # 选取六个自然月前的所有数据
    
    try:
        previousPx = stockSlice.S_DQ_ADJCLOSE[-1] # 选取离六个自然月前数据中的最后一个价格
        ret = (currentPx - previousPx) / previousPx
    except:
        ret = np.nan # 如果没有数据的话return nan，最后把这些过滤掉
        
    return ret
```

```python
%%time
df4.apply(get_return, axis=1)
```

```python
df4['ret'] = df4.apply(get_return, axis=1)
```
