# Introduction to Matplotlib
Matplotlib is a Python 2D ands 3D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. Matplotlib can be used in Python scripts, the Python and IPython shell and the jupyter notebook.

```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

plt.style.use('bmh') 
```

```python
# If your plot does not show then you must have forgotten this line
%matplotlib inline
```

## Simple Plots

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
cx, sx = np.cos(x), np.sin(x)
```

```python
plt.plot(x, cx)
plt.plot(x, sx)
```

We can change the size of the figure using a tuple

```python
plt.figure(figsize=(8,6), dpi=80)
plt.plot(x, cx)
plt.plot(x, sx)
```

We can change the colour of the lines, their thicknesses, add a label and add a legend

```python
plt.figure(figsize=(8,6), dpi=80)
plt.plot(x, cx, color='blue', label="Cos(x)", linewidth=2)
plt.plot(x, sx, color='red', label="Sin(x)", linewidth=2)
plt.title("Trigonometric Functions")
plt.legend()
```

## Advanced Customisation

Move the axes to the middle

```python
plt.figure(figsize=(8,6), dpi=80)
plt.plot(x, cx, color='blue', label="Cos(x)", linewidth=2)
plt.plot(x, sx, color='red', label="Sin(x)", linewidth=2)
plt.title("Trigonometric Functions")
plt.legend()

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

## Subplots

```python
plt.figure(figsize=(8,6), dpi=80)

plt.subplot(2, 1, 1)
plt.plot(x, sx)
plt.ylabel('Sin(x)')

plt.title('Trigonometric Functions')

plt.subplot(2, 1, 2)
plt.plot(x, cx)
plt.ylabel('Cos(x)')

plt.xlabel('Degrees (radians)')
```

## Bar Charts

```python
n = 12
x = np.arange(n)
y1 = np.random.uniform(0.0,1.0,n)
y2 = np.random.uniform(0.0,1.0,n)

plt.figure(figsize=(8,6), dpi=80)
plt.bar(x, +y1, facecolor='green', alpha=0.4, label="Y1")
plt.bar(x, +y2, facecolor='red', alpha=0.4, label = "Y2")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.bar(x, +y1, facecolor='green', alpha=0.4, label="Y1")
plt.bar(x, +y2, bottom = y1, facecolor='red', alpha=0.4, label = "Y2")
plt.xlabel("Month")
plt.ylabel("Value")
plt.xticks(x, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
plt.legend()
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.barh(x, +y1, facecolor='green', alpha=0.4, label="Y1")
plt.barh(x, +y2, left = y1, facecolor='red', alpha=0.4, label = "Y2")
plt.xlabel("Value")
plt.ylabel("Month")
plt.yticks(x, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
plt.legend()
```

## Histograms


```python
n = np.random.randn(10000)
plt.figure(figsize=(8,6), dpi=80)
plt.hist(n, bins=50, alpha=0.5, normed=1)
plt.title("Random Number Frequency Plot")
plt.xlim((min(n), max(n)))
plt.xlabel("X")
plt.ylabel("PDF")
plt.grid(True)
plt.xlim((min(n), max(n)));
```

```python
# We can calculate the cumulative histogram
n = np.random.randn(10000)
plt.figure(figsize=(8,6), dpi=80)
plt.hist(n, bins=50, alpha=0.5, normed=1, cumulative=True)
plt.title("Random Number Frequency Plot")
plt.xlim((min(n), max(n)))
plt.xlabel("X")
plt.ylabel("PDF")
plt.grid(True)
plt.xlim((min(n), max(n)));
```

```python
# Generate two vectors of data 
y = np.random.standard_normal((1000, 2))
plt.figure(figsize=(8, 6), dpi=80)
plt.hist(y, label=['1st', '2nd'], normed=True, bins=20)
plt.grid(True)
plt.legend()
plt.xlabel('X')
plt.ylabel('PDF')
```

## Scatterplots

Scatterplots can be informative about joint distributions

```python
y = np.random.standard_normal((1000, 2))
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
from matplotlib import colors
colours = ['red','green','blue']
c = np.random.randint(0, len(colours), len(y))
cmap = colors.ListedColormap(colours)
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(y[:, 0], y[:, 1], c=c, marker='o', cmap=cmap, alpha=0.6)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
plt.grid(True)
```

## Contour Plots
A contour line (also isocline, isopleth, isarithm, or equipotential curve) of a function of two variables is a curve along which the function has a constant value, so that the curve joins points of equal value.

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n) ; y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)

plt.figure(figsize=(8, 6), dpi=80)
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=0.5)
plt.clabel(C, inline=1, fontsize=10)
```

## Box Plots
In descriptive statistics, a box plot or boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending vertically from the boxes (whiskers) indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram. Outliers may be plotted as individual points. 

```python
from matplotlib import pyplot as plt
y = np.random.standard_normal((10000))
plt.figure(figsize=(8, 6), dpi=80)
f = plt.boxplot(y);

q0 = np.percentile(y,0)
q1 = np.percentile(y,25)
q2 = np.percentile(y,50)
q3 = np.percentile(y,75)
q4 = np.percentile(y,100)

iqr = (q3-q1)
l = q1 - 1.5 * iqr
h = q3 + 1.5 * iqr

print("Min: ",q0)
print("Low Bar: ",l)
print("1st quartile: ",q1)
print("Mean: ",q2)
print("3rd quartile: ",q3)
print("High Bar: ",h)
print("Max: ",q4)
```

Boxplots are useful for examining different distributions

```python
import matplotlib.pyplot as plt

N = 1000
data_set1 = np.random.normal(1, 1, N)
data_set2 = np.random.lognormal(1, 1, N)
data_set3 = np.random.exponential(1, N)
data_set4 = np.random.gumbel(6, 4, N)
data_set5 = np.random.triangular(2, 9, 11, N)

data = [data_set1, data_set2, data_set3, data_set4, data_set5]
plt.boxplot(data);
```

# 3D Plotting and Saving Plot to File

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n) ; y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
Z=f(X,Y)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0.5,antialiased=True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, shrink=0.75, aspect=4);

plt.savefig('./plots/surfaceplot.png')
plt.savefig('./plots/surfaceplot.jpeg')
plt.savefig('./plots/surfaceplot.pdf')
```
